# Memilih Warna

**Memilih value untuk argument color bisa dengan berbagai bentuk format warna, seperti format RGB, hex code, atau label warna biasa seperti 'r' untuk red, 'b' untuk blue dan lain sebagainya. Dapat dilihat pada code berikut:**

```python
import matplotlib.pyplot as plt
```

```python
# membuat figure dan axes
fig, ax = plt.subplots()

# membuat data
data_x = [1,2,3,4,5,6,7,8]
data_y = [21,42,2,3,5,12,4,10]

# memberikan data ke axes
ax.plot(data_x, data_y,marker="D",linestyle='dotted',color='#9934FF',ms=10)

# menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(12,8))
x= np.linspace(0,10,100)
plt.plot(x,np.sin(x), color='green')# ,enggunakan nama dari warna
plt.plot(x, np.sin(x-1), color='m') # menggunakan format rgbcmyk
plt.plot(x, np.sin(x-2), color='0.5') # menggunakan format graysclae (0 1)
plt.plot(x, np.sin(x-3), color='#FF0037') # menggunakan format RRGGBB,value from 00 to FF
plt.plot(x, np.sin(x-4), color=(0,1,0.5)) # format (R,G,B) 0-1
plt.plot(x, np.sin(x-5), color='DarkGreen') # HTML color names support

plt.show()
```

# Custom Axis Label dan Title

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

data_x = [1,2,3,4,5,6,7]
data_y = [23,43,31,32,36,29,15]

ax.plot(data_x,data_y, marker='x', linestyle='--', color='b')

# mengubah axis label dan tile
ax.set_xlabel('hari')
ax.set_ylabel('suhu')
# plt.title('Suhu Harian')
ax.set_title('Suhu Harian')

plt.show()
```

# Axis Limit

**Kita bisa membatasi axis dari data kita dengan menggunakan atribut xlim, ylim maupun atribut .axis, seperti berikut:**

```python
plt.figure(figsize=(8,6))
plt.plot(x,x+1,linestyle='solid')
plt.plot(x,x+2,linestyle='dashed')
plt.show()
```

**# menggunakan method .xlim dan .ylim**

```python
# menggunakan method .xlim dan .ylim
plt.figure(figsize=(10,8))
plt.plot(x,x+1,linestyle='solid')
plt.plot(x,x+2,linestyle='dashed')
plt.xlim(2,8)
plt.ylim(2,10)
plt.show()
```

**# menggunakan .axis**

```python
plt.figure(figsize=(10,8))
plt.plot(x,x+1,linestyle='solid')
plt.plot(x,x+2,linestyle='dashed')
plt.axis([1.5,8.5,2,10]) # [xmin xmax ymin ymax]
plt.show()
```

# Marker Data Point

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membuat component figure dan axes 
fig, ax = plt.subplots()

data_x = [1,2,3,4,5]
data_y = [9,15,30,12,5]

plt.plot(data_x,data_y, marker='o')

plt.show()
```

```python
plt.figure(figsize=(10,8))
x = np.random.RandomState(0)
list_marker = ['o','p','v','+','x']
for marker in list_marker:
    plt.plot(x.rand(5),x.rand(5),marker=marker,linestyle='None', label=marker,ms=15)
    
plt.legend(loc=9)
plt.show()
```

# Tutorial Video Sanbercode

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,8))
plt.plot(x,np.sin(x),label='Sin x', color = 'green')
plt.plot(x,np.cos(x),label='Cos x', color = 'red')
plt.legend(loc=9, title='legenda')
plt.title('Sinus Dan Cosinus', size=16)
plt.show()
```

**Cara Eksplisit**

```python
fig, ax = plt.subplots(sharex='all')
ax.plot(x,np.sin(x),label='sin x')
ax.plot(x,np.cos(x),label='cos x')
ax.set_ylabel('amplitude')
ax.set_xlabel('time')
ax.set_title('Sinus dan Cosinus')
plt.legend()
plt.show()
```
