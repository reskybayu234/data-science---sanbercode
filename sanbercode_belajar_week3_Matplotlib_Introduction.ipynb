**Sekarang kita akan tambahkan data kedalam objek tersebut dengan memanggil method plot() dan memberikan data sebagai argument method tersebut:**

```python
# membuat objek figure
fig = plt.figure()

# membuat objek axes
axes = fig.add_subplot()

# membuat data
data_x = [1,2,3,4,5]
data_y = [10,20,25,40,15]

# menambahkan dalam ke objek axes
axes.plot(data_x,data_y)

# menampilkan data
plt.show()
```

**Begitulah cara sederhana untuk membuat objek visualisasi, sekarang akan di tunjukan cara lain yang lebih clean.**

### cara kedua 

```python
# membuat objek visual
plt.plot(data_x,data_y)

# menampilkan objek visual
plt.show()
```

**kita juga dapat membuat cara visualisasi data lebih simple dengan menggabungkan pembuatan objek Figure dan Axes dengan sekali jalan, melalui pemanggilan method subplots(), seperti ini:**

```python
# membuat objek figure dan axes
fig, axes = plt.subplots()

# membuat data
data_x = [1,2,3,4,5]
data_y = [10,20,25,30,15]

# menambahkan data ke objek axes
axes.plot(data_x,data_y)

# menampilkan data
plt.show()
```

**Memungkinkan untuk memplot lebih dari satu data kedalam suatu axes dengan cara seperti ini:**

```python
fig, axes = plt.subplots()
data_x = [1,2,3,4,5]
data_y = [34,56,21,77,34]
data_x2 = [1,2,3,4,5]
data_y2 = [21,33,44,11,3]
axes.plot(data_x,data_y)
axes.plot(data_x2,data_y2)
plt.show()
```

# Menyimpan dan menampilkan gambar hasil plotting 

```python
# membuat figure dan axis

fig, axes = plt.subplots()
data_1 = [1,2,3,4,5]
data_2 = [5,10,5,15,20]
data_21 = [1,2,3,4,5]
data_22 = [1,7,1,14,22]
axes.plot(data_1,data_2)
axes.plot(data_21,data_22)
plt.show()

# menyimpan hasil figure
fig.savefig('coba.png')

# menampilkan figure yang disimpan
from IPython.display import Image
Image('coba.png')
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
data_x1 = [1,2,3,4,5]
data_y1 = [10,20,25,30,35]
data_x2 = [1,2,3,4,5]
data_y2 = [5,15,20,25,30]
```

**Cara Eksplisit**

```python
fig, axes = plt.subplots(figsize=(7,5))
axes.plot(data_x1,data_y1)
axes.plot(data_x2,data_y2)

# apabila ingin menambah style
plt.style.use('seaborn')

plt.show()
```

**Cara Implisit**
