```python
import matplotlib.pyplot as plt
import numpy as np
```

# with plt.axes()

```python
plt.figure(figsize=(12,8))
x = np.linspace(0,10,20) 
ax = plt.axes() 
ax2 = plt.axes((0.67,0.67,0.2,0.2)) # [left, bottom, width, height]
ax.plot(x,x+1)
ax2.plot(x,np.cos(x))
```

# with fig.add_axes()

```python
x = np.arange(10)
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_axes([0.1,0.5,0.8,0.4])
ax2 = fig.add_axes([0.1,0.1,0.8,0.4])
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x))
plt.show()
```

# with plt.subplot()

**_plt.subplot()_ digunakan untuk membuat multiple plots yang memiliki weight dan height yang sama**

```python
plt.subplot(1,2,1) # baris 1, kolom 2, posisi ke 1
plt.text(0.5,0.5, str((1,2,1)), fontsize=18,ha='center')
plt.subplot(1,2,2) # baris 1, kolom 2, posisi ke 2
plt.text(0.5,0.5,str((1,2,2)), fontsize=18, ha='center')
```

```python
plt.figure(figsize=(12,8))
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)), fontsize=18,ha='center')
```

### Sekarang mari coba membuat visualisasi data 4 plots dengan multiple axis menggunakan plt.subplots(). 

```python
import matplotlib.pyplot as plt
```

```python
data_x1 = [1,2,3,4]
data_y1 = [10,20,25,30]
data_x2 = [1,2,3,4,5]
data_y2 = [20,25,20,40,50]
data_x3 = data_x2
data_y3 = [50,10,30,15,30]
data_x4 = data_x2
data_y4 = [10,40,24,30,29]

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax[0][0].plot(data_x1, data_y1)
ax[0][1].plot(data_x2, data_y2)
ax[1][0].plot(data_x3, data_y3)
ax[1][1].plot(data_x4, data_y4)


plt.show()
```

**Coba perhatikan bahwa axis-y di setiap axis mempunyai interval yang berbeda-beda, begitu pula axis-x nya pada axis baris pertama column pertama hanya pada interval 1-4, sedangkan yang lainnya 1-5. Visualisasi ini bisa di improve agar mendapatkan interval axis yang sama dengan menambahkan argument sharex dan sharey terhadap method subplots. seperti ini:**

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
ax[0][0].plot(data_x1, data_y1)
ax[0][1].plot(data_x2, data_y2)
ax[1][0].plot(data_x3, data_y3)
ax[1][1].plot(data_x4, data_y4)

plt.show()
```

# with inset_axes()

**inset_axes adalah suatu metode dalam Matplotlib.pyplot yang digunakan untuk membuat axes di dalam sebuah axes. Seperti contoh berikut:**

```python
fig, ax = plt.subplots(figsize=(8,7))
x = np.linspace(0,10,20)
ax.plot(x,np.sin(x))

inset_cos = ax.inset_axes([0.9,0.9,0.3,0.3])
inset_cos.plot(x,np.cos(x))

inset_tan = ax.inset_axes([-0.1,0.9,0.3,0.3])
inset_tan.plot(x,np.tan(x))
```

# GridSpec()

**GridSpec digunakan untuk menspesifikasikan geometry Grid dari subplot. Seperti contoh:**

```python
grid = plt.GridSpec(2,3, wspace=0.3, hspace=0.3)

plt.figure(figsize=(12,8))
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.text(0.5,0.5, str(('in')), fontsize=18,ha='center')
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])

plt.show()
```

```python
data_x1 = [1,2,3,4]
data_y1 = [11,22,33,11]
data_x2 = [1,2,3,4,5]
data_y2 = [12,43,22,1,12]
data_x3 = data_x2
data_y3 = [23,1,55,23,87]
data_x4 = data_x2
data_y4 = [7,41,39,22,10]

fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
ax[0][0].plot(data_x1, data_y1, color='green', label=1)
ax[0][1].plot(data_x2, data_y2, color='red', label=2)
ax[1][0].plot(data_x3, data_y3, color='yellow', label=3)
ax[1][1].plot(data_x4, data_y4, color='blue', label=4)

# menambahkan legend
fig.legend()


# menambahkan title
ax[0][0].title.set_text('Hijau')
ax[0][1].title.set_text('Merah')
ax[1][0].title.set_text('Kuning')
ax[1][1].title.set_text('Biru')
plt.show()
```
