<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

# Cara 1

```python
#code here
fig, (ax1,ax2, ax3) = plt.subplots(1,3, figsize=(19,8), sharex=True, sharey=True)
ax1.plot(x,tan, color='blue', label='tanX')
ax2.plot(x,cos, color='red', label='cosX')
ax3.plot(x,sin, color='green', label='sinX')
ax1.title.set_text('Grafik tanx')
ax2.title.set_text('Grafik cosx')
ax3.title.set_text('Grafik sinx')
fig.legend(loc=9)
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(10,8), sharex=True, sharey=True)
ax.plot(x,y)
ax.set_title('Grafik cosX')

# CosX^5
inset_cos_y5 = ax.inset_axes([-0.14,0.8,0.4,0.3])
inset_cos_y5.plot(x,y5,color='yellow')
inset_cos_y5.plot(x,y,color='blue')
inset_cos_y5.title.set_text('Grafik CosX^5')

# CosX^4
inset_cos_y4 = ax.inset_axes([0.86,0.82,0.4,0.3])
inset_cos_y4.plot(x,y4,color='black')
inset_cos_y4.plot(x,y, color='blue')
inset_cos_y4.title.set_text('Grafik CosX^4')

# CosX^3
inset_cos_y3 = ax.inset_axes([0.86,-0.1,0.4,0.3])
inset_cos_y3.plot(x,y3,color='green')
inset_cos_y3.plot(x,y,color='blue')
inset_cos_y3.set_title('Grafik CosX^3', fontsize=10)

# CosX^2
inset_cos_y2 = ax.inset_axes([-0.14,-0.1,0.4,0.3])
inset_cos_y2.plot(x,y2,color='red')
inset_cos_y2.plot(x,y,color='blue')
inset_cos_y2.set_title('Grafik CosX^2', fontsize=10)



plt.show()
```
